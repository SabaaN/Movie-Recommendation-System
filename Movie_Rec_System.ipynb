```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
credits = pd.read_csv('credits.csv')
```

```python
movies.sample(5)
```

```python
credits.sample(5)
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.head(1)
```

```python
#including the columns genre, id, keywords, title, overview, cast, crew
```

```python
movies = movies [['genres' , 'id', 'title', 'overview', 'keywords' ,'cast', 'crew' ]]
```

```python
movies.head(1)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[1].genres
```

```python
import ast
```

```python
def List(obj):
    Listt = []
    for i in ast.literal_eval(obj):    #ast.literal_eval turns the string like the above into a list so that we can loop through the list and get the values
        Listt.append(i['name'])
    return Listt
```

```python
movies['genres'] = movies['genres'].apply(List)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(List)
```

```python
movies.head(1)
```

```python
movies['cast'][0]
```

```python
def actor(obj):
    Listt = []
    counter = 0
    for i in ast.literal_eval(obj): 
        if counter !=3:
            Listt.append(i['name'])  
            counter+=1
        else:
            break
    return Listt
```

```python
movies['cast'] = movies['cast'].apply(actor)
```

```python
movies.head(1)
```

```python
movies['crew'][0]
```

```python
def director(obj):
    Listt = []
    for i in ast.literal_eval(obj): 
        if i['job'] == 'Director':
            Listt.append(i['name'])  
            break
    return Listt
```

```python
movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())  #to turn the overview column into a list
```

```python
movies.head(1)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", '') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", '') for i in x])
```

```python
#we eliminated all the spaces between one single name so that it remains a single entity instead of becoming multiple entities
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']  
```

```python
movies.head(1)
```

```python
movies_new = movies[['id', 'title', 'tags']]
```

```python
movies_new.head(15)
```

```python
movies_new['tags']  = movies_new['tags'].apply(lambda x:" ".join(x))  #converting the tags column from list to string
```

```python
movies_new.head(5)
```

```python
movies_new['tags']  = movies_new['tags'].apply(lambda x:x.lower()) #converting tags column to lowercase
```

```python
movies_new.head(5)
```

```python
#half of preprocessing(sorting out the data) is done here. Now begins the vectorization and the actual recommendation system building
#we will be converting our text into vectors (each movie will be converted into one vector and then use bag of words technique for text vectorization)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english',  max_features=5000)
```

```python
vec = cv.fit_transform(movies_new['tags']).toarray()
```

```python
vec[0]
```

```python
cv.get_feature_names_out().tolist()
```

```python
import nltk
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stemming(obj):
    str = []
    for i in obj.split():
        str.append(ps.stem(i))
    return ' '.join(str)
```

```python
movies_new['tags'] = movies_new['tags'].apply(stemming)
```

```python
cv.get_feature_names_out().tolist()
```

```python
#now we will be calculating the cosine distance(i.e angle between two vectors) the larger the data, the more euclidean distance fails
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vec)
```

```python
similar[0]
```

```python
def recs(movie):
    if movie in movies_new['title'].values:
        movieIndex = movies_new[movies_new['title'] == movie].index[0]
        dist = similar[movieIndex]
        movieList = sorted(list(enumerate(dist)), reverse = True, key = lambda x:x[1])[1:7]
    
        for i in movieList:
            print(movies_new.iloc[i[0]].title)
    else:

        print(f"Uh oh! Looks like {movie} not found in the movie database:(")
    
    
```

```python
recs('Batman')
```
